Цель работы: научится работать с библиотеками matplotlib и pandas, проанализировать датасет по героям игры "LOL".

```python
import matplotlib.pyplot as plt
import pandas as dp
import re
tbl = dp.read_csv('LOL_champions_stats.csv', encoding="ISO-8859-1", delimiter=';')
```

```python
x = tbl['champion_name'].tail(100)
y = tbl['blue_essence_price'].tail(100)

plt.figure(figsize=(20, 5))
plt.bar(x, y, label='Стоимость')
plt.title('Стоимость чемпионов в синих эсенциях')
plt.xlabel('Чемпион')
plt.ylabel('Стоимость')
plt.legend()
plt.show()
```

Данная столбчатая диаграмма демонстрирует цены чемпионов. Здесь четко видно, что существуют повторяющиеся, типовые цены.

```python
x = tbl['champion_name'].tail(100)
y = tbl['rp_price'].tail(100)

plt.figure(figsize=(20, 5))
plt.bar(x, y, label='Стоимость')
plt.title('Стоимость чемпионов в Riot points')
plt.xlabel('Чемпион')
plt.ylabel('Riot points')
plt.legend()
plt.show()
```

Здесь мы также видим преимущество типовых цен чемпионов в Riot points.

```python
be = tbl['blue_essence_price'].tail(20)
rp = tbl['rp_price'].tail(20)
be_avr = sum(be) / len(be)
rp_avr = sum(rp) / len(rp)

plt.pie([be_avr, rp_avr])
plt.title('Соотношение стоимости чемпионов (20)')
plt.legend()
plt.show()
```

Соотношение средней стоимости чемпионов в синих эссенциях и Riot points с выборкой 20 чемпионов.

```python
be = tbl['blue_essence_price'].tail(100)
rp = tbl['rp_price'].tail(100)
be_avr = sum(be) / len(be)
rp_avr = sum(rp) / len(rp)

plt.pie([be_avr, rp_avr])
plt.title('Соотношение стоимости чемпионов (100)')
plt.legend()
plt.show()
```

Соотношение средней стоимости чемпионов в синих эссенциях и Riot points с выборкой 100 чемпионов. Так как последние 2 круговые диаграммы не имеют различий, можно сделать вывод, что соотношение между стоимостями чемпионов в разных валютах одинаково, а значит не существует более выгодной валюты для покупки. Баланс выстроен верно.

```python
def extract_float(s):
    match = re.search(r'\d+\.\d+', s)
    if match:
        return float(match.group())
    else:
        return None
x = tbl['champion_name'].tail(100)
y = [extract_float(s) for s in tbl['last_change'].tail(100)]

plt.figure(figsize=(20, 5))
plt.bar(x, y, label='Последняя версия изменений')
plt.title('Последняя версия изменений чемпионов')
plt.xlabel('Чемпион')
plt.ylabel('Версия изменения')
plt.legend()
plt.show()
```

Исходя из этого графика можно сказать, что чемпионов обновляют часто, так как при выходе новых необходимо переделывать баланс игры под старых, для того, чтобы все персонажи были примерно равны по силе или полезности.

```python
data = tbl['HP'].tail(158)

plt.figure(figsize=(10, 5))
plt.hist(data, bins=20)
plt.title('HP чемпионов')
plt.xlabel('Чемпион')
plt.ylabel('HP')
plt.legend()
plt.show()
```

Из этой гитограммы можно увидеть, что среднее количество HP чемпиона равно около 550-600, при этом значения низкие и высокие значения встречаются реже всего.

```python
data = tbl['MP'].tail(158)

plt.figure(figsize=(10, 5))
plt.hist(data, bins=20)
plt.title('MP чемпионов')
plt.xlabel('Чемпион')
plt.ylabel('MP')
plt.legend()
plt.show()
```

По данной гистограмме можно понять, среднее что количество манны у чемпионов равняется примерно 300, при этом количество чемпионов с низкими и высокими значениями манны примерно равно. 

```python
data = tbl['AD'].tail(158)

plt.figure(figsize=(10, 5))
plt.hist(data, bins=20)
plt.title('AD чемпионов')
plt.xlabel('Чемпион')
plt.ylabel('AD')
plt.legend()
plt.show()
```

На данной гистограмме можно увидеть, что атака чемпионов постепенно возрастает пока не достигает пика в 60-65, после чего резко до падает низа - 70.

```python
x = tbl['champion_name'].tail(100)
y1 = tbl['AD'].tail(100)
y2 = tbl['AD+'].tail(100)
y3 = tbl['AD_lvl18'].tail(100)

plt.figure(figsize=(20, 5))
plt.plot(x, y1, label='Атака', color='blue')
plt.plot(x, y2, label='Атака+', color='green')
plt.plot(x, y3, label='Атака 18 уровень', color='red')
plt.title('Зависимости значений атаки')
plt.xlabel('Чемпион')
plt.ylabel('Атака')
plt.legend()
plt.show()
```

Этот линейный график отображает зависимости 3 параметров атаки. Здесь видно что все графики хоть и различаются, но они немного совпадают, за исключением нескольких аномальных значений.

```python
x = tbl['MP'].tail(100)
y = tbl['AD'].tail(100)


plt.figure(figsize=(10, 5))
plt.scatter(x, y)
plt.title('Зависимости значений атаки и манны')
plt.xlabel('MP')
plt.ylabel('AD')
plt.legend()
plt.show()
```

В этой диаграмме рассеивания видно, что для чемпионов с более большим запасом характерна более низкая атака.

```python
x = tbl['MP'].tail(100)
y = tbl['HP'].tail(100)


plt.figure(figsize=(10, 5))
plt.scatter(x, y)
plt.title('Зависимости значений здоровья и манны')
plt.xlabel('MP')
plt.ylabel('HP')
plt.legend()
plt.show()
```

В данной диаграмме рассеивания видно, что значения здоровья мало зависят от запаса манны.
